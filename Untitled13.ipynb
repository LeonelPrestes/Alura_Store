# Relatório Final de Análise de Churn - TelecomX

## Introdução
O objetivo principal desta análise é investigar e compreender os fatores que influenciam a evasão de clientes (Churn) na TelecomX. Através da exploração de dados demográficos, serviços assinados e informações financeiras, buscamos identificar padrões que permitam à empresa tomar decisões estratégicas para aumentar a retenção de seus usuários.

## Tratamento de Dados
Para garantir a qualidade das análises, os dados passaram por um rigoroso processo de limpeza e preparação, que incluiu as seguintes etapas:

1. **Carga de Dados**: Importação do conjunto de dados original no formato JSON a partir de um repositório remoto.
2. **Expansão de Colunas Aninhadas**: As informações estruturadas nos campos `phone`, `internet` e `account` foram expandidas (achatadas) utilizando `json_normalize` para transformar dicionários em colunas individuais.
3. **Tratamento de Valores Ausentes**: Identificamos valores nulos na coluna `Charges.Total` (Cobrança Total) após a conversão para numérico, que foram preenchidos com a média da própria coluna.
4. **Padronização de Categorias**: Simplificamos as variáveis categóricas, substituindo termos redundantes como 'No phone service' e 'No internet service' por 'No', e tratando entradas vazias na coluna de Churn.
5. **Conversão e Preparação**: Colunas financeiras foram convertidas para tipos numéricos (float) e variáveis qualitativas foram padronizadas para facilitar a análise visual e a codificação posterior.

### Análise de Insights

Com base nas visualizações geradas, podemos identificar padrões claros de evasão (churn) entre os clientes da TelecomX:

1.  **Tipo de Contrato**: O gráfico de churn por contrato revela que clientes com contratos **'Month-to-month' (mês a mês)** possuem a maior taxa de evasão (42.7%). Em contraste, clientes com contratos de um ou dois anos apresentam uma fidelidade significativamente maior, sugerindo que a flexibilidade do contrato mensal facilita a saída do cliente.

2.  **Serviço de Internet**: Existe uma discrepância notável entre as tecnologias de internet. Clientes que utilizam **'Fiber optic' (fibra ótica)** têm uma taxa de churn muito superior (41.9%) em comparação aos usuários de DSL (19.0%) ou sem internet. Isso pode indicar problemas de preço ou instabilidade técnica especificamente no serviço de fibra.

3.  **Método de Pagamento**: O método de pagamento **'Electronic check' (cheque eletrônico)** está fortemente correlacionado com a evasão, apresentando a maior proporção de cancelamentos entre todas as opções de pagamento (45.3%). Métodos automáticos, como cartão de crédito ou transferência bancária, mostram taxas de churn consideravelmente menores.

4.  **Serviço de Telefone**: A presença ou ausência de serviço de telefone parece ter um impacto marginal na decisão de evasão, com taxas de churn similares (aproximadamente 25% para ambos os grupos).

**Conclusão e Perfil de Risco:**
O perfil do cliente com maior propensão à evasão é aquele que possui um **contrato mensal**, utiliza **fibra ótica** como serviço de internet e realiza pagamentos via **cheque eletrônico**. Estes dados sugerem que estratégias de retenção devem focar na migração desses clientes para planos de longo prazo e incentivar o uso de métodos de pagamento automáticos.

### Recomendações Estratégicas

Com base na análise dos dados de evasão (churn) da TelecomX, seguem as recomendações estratégicas para retenção de clientes:

1. **Incentivo à Migração de Contratos**: Os dados mostram uma taxa de churn significativamente maior em contratos mensais (*Month-to-month*). Recomenda-se a implementação de campanhas de fidelidade com descontos progressivos ou benefícios exclusivos para clientes que migrarem para contratos de um ou dois anos.

2. **Promoção de Métodos de Pagamento Automáticos**: Clientes que utilizam cheque eletrônico (*Electronic check*) apresentam maior propensão à evasão. Sugere-se oferecer bônus na fatura ou pacotes de dados extras para clientes que cadastrarem métodos de pagamento automáticos (Cartão de Crédito ou Transferência Bancária), garantindo uma cobrança mais fluida e menor atrito.

3. **Auditoria e Melhoria do Serviço de Fibra Ótica**: Identificamos que a tecnologia de Fibra Ótica apresenta um churn elevado em comparação ao DSL. É fundamental realizar uma auditoria técnica e pesquisas de satisfação específicas com esses usuários para entender se há problemas de instabilidade de sinal ou se o preço não está competitivo com a qualidade entregue.

4. **Programas de Suporte Proativo**: Como o suporte técnico e a segurança online mostraram ser diferenciais, a empresa deve empacotar esses serviços como cortesia para novos assinantes de fibra, aumentando o valor percebido e a barreira de saída do cliente.
