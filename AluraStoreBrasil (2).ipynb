### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
precos = loja['Preço']
precos2 = loja2['Preço']
precos3 = loja3['Preço']
precos4 = loja4['Preço']

faturamento = float(precos.sum())
faturamento2 = float(precos2.sum())
faturamento3 = float(precos3.sum())
faturamento4 = float(precos4.sum())

print(faturamento)
print(faturamento2)
print(faturamento3)
print(faturamento4)
```

## Soma dos fretes por loja

```python
somaFrete = loja['Frete'].sum()
somaFrete2 = loja2['Frete'].sum()
somaFrete3 = loja3['Frete'].sum()
somaFrete4 = loja4['Frete'].sum()

print(f'Loja 1 \n{somaFrete}\n')
print(f'Loja 2 \n{somaFrete2}\n')
print(f'Loja 3 \n{somaFrete3}\n')
print(f'Loja 4 \n{somaFrete4}\n')
```

## Faturamento sem frete por loja

Sabemos que o faturamente de cada unidade é:
1. loja 1 = 1534509.12
2. loja 2 = 1488459.06
3. loja 3 = 1464025.03
4. loja 4 = 1384497.58  

O valor do frete é pago pelo cliente, porém devemos descontar do valor de faturamento bruto, pois esse valor é repassado para a transportadora.
Então abatendo o valor do frente do faturamento bruto temos...
1. loja 1 = 1452671.1522043084
2. loja 2 = 1409144.9019873552
3. loja 3 = 1386002.368777357
4. loja 4 = 1310741.7048810173  

Então a loja com maior faturamento, descontando o frete, é a loja 1, e a com menor faturamente é a loja 4.

```python
loja1FaturamentoSemFrete = faturamento - somaFrete
loja2FaturamentoSemFrete = faturamento2 - somaFrete2
loja3FaturamentoSemFrete = faturamento3 - somaFrete3
loja4FaturamentoSemFrete = faturamento4 - somaFrete4

print(loja1FaturamentoSemFrete)
print(loja2FaturamentoSemFrete)
print(loja3FaturamentoSemFrete)
print(loja4FaturamentoSemFrete)
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [loja1FaturamentoSemFrete, loja2FaturamentoSemFrete, loja3FaturamentoSemFrete, loja4FaturamentoSemFrete]

plt.figure(figsize=(8, 5))
plt.bar(lojas, valores)

plt.title('Faturamento descontando o frete')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.6)

for i, v in enumerate(valores):
    plt.text(i, v + 100, f'R$ {v:,.2f}', ha='center', fontsize=10)

plt.show()
```
